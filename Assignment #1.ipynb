````python
time_series = pd.read_csv("C:/Users/Hamza Amjad/Desktop/Machine Learning/Assignment #1/time_series_covid_19_deaths.csv")

# Examine the data
count = time_series.isnull().sum()
print(f"Data that contain null values\n{count[0:8]}")

key = "Deaths_Report"
time_series = time_series.dropna()
count = time_series.isnull().sum()
print(f"\nCalculating Null Values\n{count[0:8]}")
time_series.head(10)
time_series2 = pd.melt(time_series, id_vars = ['Province','Country','Lat','Long'], var_name='Dates', value_name='Deaths_Report')
time_series2['Dates'] = pd.to_datetime(time_series2['Dates'], infer_datetime_format=True)

df_filtered1 = time_series2.query('Country=="US"')
data6 = df_filtered1.sort_values(by=key, ascending=False)
most_affected_states1 = data6.groupby(['Province']).sum()
most_affected_states1 = most_affected_states1.sort_values(by=key,ascending=False)
X = most_affected_states1.head(5)
Y = np.array(most_affected_states1.head(5).iloc[:,2])
X = X.index[:5]
figure(figsize = (10, 8), dpi=60)
plt.title("US Death Graph")
plt.xlabel("States of US")
plt.ylabel("Death Cases")
plt.bar(X, Y, width=0.9, color="blue")
plt.show()

    
df_filtered2 = time_series2.query('Country=="China"')
data7 = df_filtered2.sort_values(by=key, ascending=False)
most_affected_states2 = data7.groupby(['Province']).sum()
most_affected_states2 = most_affected_states2.sort_values(by=key,ascending=False)
B=np.array(most_affected_states2.head(5).iloc[:,2])
A = most_affected_states2.head(5).index[:5]
figure(figsize=(10, 8), dpi=60)
plt.title("China Death Graph")
plt.xlabel("States of China")
plt.ylabel("Death Cases")
plt.bar(A, B, width=0.9, color="red")
plt.show()
````

````python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np


"""Step 1: Load world map and plot values on map"""

df_covid = pd.read_csv('time_series_covid_19_deaths.csv')

fig = plt.figure(figsize=(12,9))
m = Basemap(
    projection='mill',
    llcrnrlat=-90,
    urcrnrlat=90,
    llcrnrlon=-180,
    urcrnrlon=180,
    resolution='c'
)
m.drawcoastlines()

m.drawparallels(np.arange(-90,90,10), labels=[True, False,False, False])
m.drawmeridians(np.arange(-180,180,30), labels=[0,0,0,1])
for idx, row in df_covid.iterrows():
  m.scatter(row['Long'], row['Lat'], latlon=True, c='red')


plt.title('Geographical Plot of COVID-19 Cases', fontsize=14)
plt.show()


"""Step 2: Visually examine the graph to find the countries where most covid cases are spotted"""

print('Visual Examination:\nFrom the graph it can be observed that most cases took place in eastern China and South-Eastern Europe.')
````
